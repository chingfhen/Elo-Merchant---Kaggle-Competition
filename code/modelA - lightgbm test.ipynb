```python
import lightgbm as lgb
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_validate
```

### LGBMRegressor

```python
path = r"C:\Users\tanch\Documents\NTU\NTU Year 4\Semester 1\CZ4041 - Machine Learning\Team Project\data\train.csv"
data = pd.read_csv(path, nrows = 100000)
data = data.sample(frac=1)
train, test = data.iloc[:80000],data.iloc[:20000]
X_cols = ['feature_1','feature_2','feature_3']
y_col = "target"
categorical_features = ['feature_1','feature_2','feature_3']
X_train,y_train = train[X_cols], train[y_col]
X_test,y_test = test[X_cols], test[y_col]

```

```python
params = {"n_estimators":1000}
model = lgb.LGBMRegressor(**params)

model.fit(X = X_train, y = y_train)
```

```python
cv = 5
scoring=('r2', 'neg_root_mean_squared_error')
verbose = 1
```

```python
scores = cross_validate(estimator = model, 
                        X = X_train, 
                        y = y_train, 
                        cv=cv,
                        scoring=scoring,
                        verbose = 10,
                        return_train_score=True)
scores["train_neg_root_mean_squared_error"].mean()
```
