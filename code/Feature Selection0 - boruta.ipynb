```python
from config import *

import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_validate
import numpy as np

from boruta import BorutaPy
```

```python
path = r"C:\Users\tanch\Documents\NTU\NTU Year 4\Semester 1\CZ4041 - Machine Learning\Team Project\data\train_df.pkl\train_node2vec_embeddings.pkl"
train_file = pd.read_csv(train_path,usecols = ['card_id', 'feature_1', 'feature_2', 'feature_3','target'])
feature_file = pd.read_pickle(path)
data = train_file.merge(feature_file, on = "card_id", how = "left")
del train_file, feature_file
data
```

```python
all_features = [col for col in data.columns if col not in ["card_id","target"]]
X, y = data[all_features], data["target"]
X
```

## boruta

```python
model_params = {
    "learning_rate":0.01,
    "n_estimators":200,
    "subsample":0.8,
    "max_depth":20,
    "n_iter_no_change":5,
    "verbose":1,
    "random_state":0
}
model = GradientBoostingRegressor(**model_params)
```

```python
boruta = BorutaPy(
    estimator = model, 
    n_estimators = 'auto',
    max_iter = 100, # number of trials to perform
    random_state = 0,
    perc = 100,
    verbose = 2,
    alpha = 0.05
)
### fit Boruta (it accepts np.array, not pd.DataFrame)
boruta.fit(np.array(X), np.array(y))
```

```python
print("Confirmed:", X.columns[boruta.ranking_==1].tolist())
print("Tentative:", X.columns[boruta.ranking_==2].tolist())
print("Rejected:", X.columns[boruta.ranking_==3].tolist())
```

## evaluate selected features

```python
selected_features = X.columns[boruta.ranking_==1].tolist()
all_features = X.columns.tolist()
target_col = "target"
results = {}
```

```python
X,y = data[all_features], data[target_col]
cv = 5
scoring=('r2', 'neg_root_mean_squared_error')
verbose = 0
model_params = {
    "learning_rate":0.01,
    "n_estimators":100,
    "subsample":0.8,
    "n_iter_no_change":5,
    "verbose":1,
    "random_state":0
}
model = GradientBoostingRegressor(**model_params)
```

```python
scores = cross_validate(estimator = model, 
                        X = X, 
                        y = y, 
                        cv=cv,
                        scoring=scoring,
                        verbose = verbose,
                        return_train_score=True)
results["all_features"] = scores
```

```python
X,y = data[selected_features], data[target_col]
model = GradientBoostingRegressor(**model_params)
```

```python
scores = cross_validate(estimator = model, 
                        X = X, 
                        y = y, 
                        cv=cv,
                        scoring=scoring,
                        verbose = verbose,
                        return_train_score=True)
results["boruta_features"] = scores
```

```python
from numpy import mean

def get_mean(results):
    output = {}
    for key,scores in results.items():
        output[key] = mean(scores)
        
    return output
```
