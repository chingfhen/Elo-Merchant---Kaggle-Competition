```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
path = r"C:\Users\tanch\Documents\NTU\NTU Year 4\Semester 1\CZ4041 - Machine Learning\Team Project\results\node2vec_embeddings.csv"
results = pd.read_csv(path)
results_long = pd.melt(results, id_vars = ["type","id_column","embedding_dimension"], value_vars = ["train_r2","test_r2","train_root_mean_squared_error","test_root_mean_squared_error"])
results_long
```

```python
def process_data(data, id_column, metrics):
    temp = data
    temp = temp[temp["id_column"] == id_column]
    temp = temp[temp["variable"].isin(metrics)]
    return temp
```

# merchant_id

```python
metric = "R_Square"
x_ticks = ['baseline','+embeddings','+id_features','+all']
temp = process_data(results_long, "merchant_id", ["train_r2","test_r2"])
g = sb.barplot(data = temp,  y = "value", hue = "variable", x = "type", palette = "Accent_r")
g.legend(title=None)
g.set_xticks([0,1,2,3]) # <--- set the ticks first
g.set_xticklabels(x_ticks)
g.set(ylabel = metric, xlabel = None)
```

```python
metric = "RMSE"
x_ticks = ['baseline','+embeddings','+id_features','+all']
temp = process_data(results_long, "merchant_id", ["train_root_mean_squared_error","test_root_mean_squared_error"])
g = sb.barplot(data = temp,  y = "value", hue = "variable", x = "type", palette = "Accent_r")
g.legend(title=None)
g.set_xticks([0,1,2,3]) # <--- set the ticks first
g.set_xticklabels(x_ticks)
g.set(ylabel = metric, xlabel = None)
g.set_ylim(temp.value.min()-temp.value.std(), temp.value.max()+ temp.value.std())
```

```python
def process_data(data, id_column, metrics, typee = None):
    temp = data
    temp = temp[temp["id_column"].isin(id_column)]
    temp = temp[temp["variable"].isin(metrics)]
    if typee:
        temp = temp[temp["type"].isin(typee)]
    return temp
```

```python
metric = "RMSE"
x_ticks = ["16","8"]
temp = process_data(results_long, ['state_id'], ["train_root_mean_squared_error","test_root_mean_squared_error"], typee = ["baseline_with_embeddings"])
g = sb.barplot(data = temp,  y = "value", hue = "variable", x = "embedding_dimension", palette = "Accent_r")
g.legend(title=None)
g.set_xticks([0,1]) # <--- set the ticks first
g.set_xticklabels(x_ticks)
g.set(ylabel = metric, xlabel = "embedding_dimension")
g.set_ylim(temp.value.min()-temp.value.std(), temp.value.max()+ temp.value.std())
```
